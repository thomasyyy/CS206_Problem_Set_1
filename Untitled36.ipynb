```python
def build_payoff_matrices(grid, cap=100):
    """Return A (P1 payoff), B (P2 payoff) for given action grid and capacity rule."""
    grid = list(grid)
    n = len(grid)
    A = np.zeros((n, n), dtype=float)
    B = np.zeros((n, n), dtype=float)
    for i, d1 in enumerate(grid):
        for j, d2 in enumerate(grid):
            if d1 + d2 <= cap:
                A[i, j] = d1
                B[i, j] = d2
            else:
                A[i, j] = 0.0
                B[i, j] = 0.0
    return grid, A, B

def validate_rule(grid, A, B, cap=100):
    """Hard checks to ensure the matrices implement the rule exactly."""
    grid = list(grid)
    n = len(grid)
    for i, d1 in enumerate(grid):
        for j, d2 in enumerate(grid):
            if d1 + d2 <= cap:
                assert A[i, j] == d1 and B[i, j] == d2, (i, j, d1, d2, A[i,j], B[i,j])
            else:
                assert A[i, j] == 0 and B[i, j] == 0, (i, j, d1, d2, A[i,j], B[i,j])
    return "OK"

```

```python
STEP = 25  # 你可以改成 10/5/1
GRID = list(range(0, 101, STEP))
grid, A, B = build_payoff_matrices(GRID, cap=100)
print("Validation:", validate_rule(grid, A, B))

dfA = pd.DataFrame(A, index=[f"P1:{x}" for x in grid], columns=[f"P2:{y}" for y in grid])
dfB = pd.DataFrame(B, index=[f"P1:{x}" for x in grid], columns=[f"P2:{y}" for y in grid])

print("Payoff matrix A (P1):")
display(dfA)
print("\nPayoff matrix B (P2):")
display(dfB)
```

```python
def pure_NE_by_characterization(grid, cap=100):
    nes=[]
    for d1 in grid:
        for d2 in grid:
            if d1 + d2 == cap:
                nes.append((d1, d2, d1, d2))  # payoffs = (d1,d2)
    return nes

nes = pure_NE_by_characterization(grid, cap=100)
print(f"Pure NE count (sum=100) on step={STEP}: {len(nes)}")
print("First 10 examples:", nes[:10])
```

```python
game = nash.Game(A, B)
eqs = list(game.support_enumeration())
print(f"NashPy found {len(eqs)} equilibria")
for k, (p, q) in enumerate(eqs, start=1):
    supp_p = [grid[i] for i,pi in enumerate(p) if pi>1e-10]
    supp_q = [grid[j] for j,qj in enumerate(q) if qj>1e-10]
    u1 = float(p @ A @ q); u2 = float(p @ B @ q)
    print(f"Eq {k}: supp(P1)={supp_p}, supp(P2)={supp_q}, E[U]=({u1},{u2})")
```

```python
for STEP_FINE in [10, 5, 1]:
    GRID_FINE = list(range(0, 101, STEP_FINE))
    nes_fine = pure_NE_by_characterization(GRID_FINE, cap=100)
    print(f"STEP={STEP_FINE}: NE count (sum=100) = {len(nes_fine)}; examples: {nes_fine[:5]} ... {nes_fine[-5:]}")
```

```python
GRID_150 = [0, 50, 100, 150]
grid150, A150, B150 = build_payoff_matrices(GRID_150, cap=100)
print("Validation (grid with 150):", validate_rule(grid150, A150, B150))

# 纯 NE 直接扫一遍（包含 (150,150)）
def pure_NE_bruteforce(grid, A, B):
    n = len(grid)
    nes=[]
    # best responses
    br_rows=[set() for _ in range(n)]
    br_cols=[set() for _ in range(n)]
    for j in range(n):
        col = A[:, j]; br_rows[j]=set(np.where(col>=col.max()-1e-12)[0])
    for i in range(n):
        row = B[i, :]; br_cols[i]=set(np.where(row>=row.max()-1e-12)[0])
    for i in range(n):
        for j in range(n):
            if i in br_rows[j] and j in br_cols[i]:
                nes.append((grid[i], grid[j], A[i,j], B[i,j]))
    return nes

nes150 = pure_NE_bruteforce(grid150, A150, B150)
print("NE on {0,50,100,150}:", nes150)
```
